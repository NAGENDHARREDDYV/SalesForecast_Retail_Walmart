```python
for key in weather.keys():
    if key not in ['Year-Month','Day']:
        weather[key] = weather[key].convert_objects(convert_numeric=True)
```

```python
weather['WeatherEvent'] = weather['WeatherEvent'].astype('str') 

col_names =  ['index','Rain', 'Snow', 'Fog', 'Year-Month']

WeatherCount  = pd.DataFrame(columns = col_names)  

for item, frame in weather['WeatherEvent'].iteritems():
            WeatherCount.at[item, 'Rain'] = 0
            WeatherCount.at[item, 'Snow'] = 0
            WeatherCount.at[item, 'Fog'] = 0
    
for item, frame in weather['WeatherEvent'].iteritems():
        if frame=='nan':
            WeatherCount.at[item, 'Rain'] = 0
            WeatherCount.at[item, 'Snow'] = 0
            WeatherCount.at[item, 'Fog'] = 0
        else :
            frame_split=frame.split(',')
            for tup in frame_split:

                if tup.strip() == 'Rain':
                    WeatherCount.at[item, 'Rain']=WeatherCount.at[item, 'Rain'] + 1
                if tup.strip()  == 'Snow':
                    WeatherCount.at[item, 'Snow']=WeatherCount.at[item, 'Snow'] + 1
                if tup.strip()  == 'Fog':
                    WeatherCount.at[item, 'Fog']=WeatherCount.at[item, 'Fog'] + 1
                    
WeatherCount['Year-Month']=weather['Year-Month']


def WeatherAgregator_mean(data):
    #data = pd.read_csv(weatherfile)
    data_mod = data.groupby(['Year-Month']).mean()
    #del data_mod["Day"]
    return data_mod  

def WeatherAgregator_sum(data):
    #data = pd.read_csv(weatherfile)
    data_mod = data.groupby(['Year-Month']).sum()
    #del data_mod["Day"]
    return data_mod  

weather_mergedmean=WeatherAgregator_mean(weather)
weather_mergedsum=WeatherAgregator_sum(WeatherCount)

# Concat the two datasets combine
weather_merged = pd.concat([weather_mergedmean,weather_mergedsum],axis=1)
del weather_merged["Day"]
weather.dtypes
del weather_merged['index']
```
